```python
# Make the hour list of 2018


# import pandas as pd
# from datetime import datetime, timedelta
# import pytz
# 
# # Define the time zone
# timezone = pytz.timezone('Europe/Amsterdam')  # Change this to the appropriate time zone
# 
# # Define the start and end dates of 2018
# start_date = datetime(2018, 1, 1)
# end_date = datetime(2019, 1, 1)
# 
# # List to store all timestamps
# hours_of_day_2018 = []
# 
# # Generate timestamps for each hour of each day in 2018
# current_date = start_date
# while current_date < end_date:
#     for hour in range(24):
#         timestamp = timezone.localize(datetime(current_date.year, current_date.month, current_date.day, hour))
#         hours_of_day_2018.append(timestamp.hour)
#     current_date += timedelta(days=1)
# 
# # Create a DataFrame with the list of integers
# df = pd.DataFrame({'Hour': hours_of_day_2018})
# 
# # Write the DataFrame to a CSV file
# df.to_csv('hours_of_day_2018_with_dst.csv', index=False)
# 
####################################################################

# import pandas as pd
# 
# # Aantal dagen in 2018
# days_in_2018 = 365  # Voor een niet-schrikkeljaar
# 
# # Lijst om de uren van 2018 op te slaan
# hours_of_2018 = []
# 
# # Loop door alle dagen in 2018
# for day in range(1, days_in_2018 + 1):
#     # Voeg 24 uur toe voor elke dag
#     for hour in range(24):
#         hours_of_2018.append(hour)  # Uren beginnen vanaf 0
# 
# # Maak een DataFrame van de lijst van uren
# df = pd.DataFrame({'Hour': hours_of_2018})
# 
# # Voeg een kolom toe met een eenvoudige teller die oploopt voor elk uur, beginnend bij 0
# df['Time'] = range(len(df))
# 
# # Sla de DataFrame op naar een CSV-bestand
# df.to_csv('hours_of_2018.csv', index=False)




# Creating Time and Hour

import pandas as pd

# Number of days in 2018
days_in_2018 = 365  # For a non-leap year

# List to store the hours of 2018
hours_of_2018 = []

# Loop through all the days in 2018
for day in range(1, days_in_2018 + 1):
    # Add 24 hours for each day
    for hour in range(24):
        hours_of_2018.append(hour)  # Hours start from 0

# Create a DataFrame from the list of hours
df = pd.DataFrame({'Hour': hours_of_2018})

# Add a column with a simple counter that increments for each hour, starting from 0
df['Time'] = range(len(df))

# Reorder the columns so that 'Time' comes before 'Hour'
df = df[['Time', 'Hour']]

# Save the DataFrame to a CSV file
df.to_csv('hours_of_2018.csv', index=False)

```

```python
# Create function to remove the first column

def copy_csv_remove_first_column(input_file, output_file):
    """
    Kopieert een CSV-bestand en verwijdert de eerste kolom voordat het wordt opgeslagen.
    
    Parameters:
        input_file (str): Het pad naar het invoer CSV-bestand.
        output_file (str): Het pad naar het uitvoer CSV-bestand.
    """
    # Lees het originele CSV-bestand in een DataFrame
    df = pd.read_csv(input_file)

    # Maak een kopie van de DataFrame zonder de eerste kolom
    df_copy = df.copy()
    df_copy = df_copy.drop(columns=[df.columns[0]])  # Verwijder de eerste kolom

    # Schrijf de DataFrame naar een CSV-bestand
    df_copy.to_csv(output_file, index=False)

# Voorbeeldgebruik van de functie
copy_csv_remove_first_column('werkdagen_2018_nl.csv', 'workingday_2018.csv')

#   --> Workingday is correct

```

```python
# Extract all the individual bus load values from modified_aggre_load and put it as separated files in a folder

import pandas as pd
import os

# Path to the original CSV file
csv_file_path = 'bap/modified_aggr_load.csv'

# Read the entire CSV file without header
data = pd.read_csv(csv_file_path, header=None)

# Create the output directory if it doesn't exist
output_dir = 'bap/data'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Loop over each column in the DataFrame, skipping the first column
for i in range(1, data.shape[1]):  # Start from 1 to skip the first column
    # Extract the column data
    column_data = data[[i]].copy()

    # Rename the column to 'Load'
    column_data.columns = ['Load']

    # Remove the first row and shift the remaining data up
    column_data = column_data.drop(index=0).reset_index(drop=True)

    # Path to the new CSV file for the current column
    output_csv_file_path = os.path.join(output_dir, f'bus_{i}_load.csv')

    # Save the column data as a separate CSV file
    column_data.to_csv(output_csv_file_path, index=False)
    print(f'Saved column {i} to {output_csv_file_path}')


```

```python
# Import all the hours time and workingdays and put it as separated files in a folder called dataset_complete

import pandas as pd
import os

# Paths to the constant CSV files
hours_of_2018_file_path = 'hours_of_2018.csv'
workingday_2018_file_path = 'bap/workingday_2018_formatted.csv'

# Read the constant CSV files
hours_of_2018 = pd.read_csv(hours_of_2018_file_path, usecols=[0, 1])
workingday_2018 = pd.read_csv(workingday_2018_file_path, usecols=[0])

# Input and output directories
input_dir = 'bap/data'
output_dir = 'bap/dataset_complete'

# Create the output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Loop over each bus_i_load.csv file in the input directory
for i in range(1, 29):  # Range from 1 to 28 inclusive
    bus_file_path = os.path.join(input_dir, f'bus_{i}_load.csv')

    # Read the current bus file
    bus_data = pd.read_csv(bus_file_path, header=None, names=['Load'])

    # Merge the datasets based on the index
    merged_dataset = pd.concat([hours_of_2018, workingday_2018, bus_data], axis=1)

    # Path to the new CSV file for the current bus
    output_csv_file_path = os.path.join(output_dir, f'bus_{i}_complete.csv')

    # Save the merged dataset as a new CSV file
    merged_dataset.to_csv(output_csv_file_path, index=False)
    print(f'Saved complete dataset for bus {i} to {output_csv_file_path}')



# define new function to shift the values:
def shift_load_column_up(file_path):
    # Read the CSV file
    df = pd.read_csv(file_path)
    
    # Shift the "Load" column up by one position and drop the last row
    df['Load'] = df['Load'].shift(-1)
    df = df.dropna().reset_index(drop=True)
    
    # Save the modified DataFrame back to the CSV file
    df.to_csv(file_path, index=False)
    print(f'Processed and saved: {file_path}')

# Directory containing the files
input_dir = 'bap/dataset_complete'

# Iterate over each file in the directory
for file_name in os.listdir(input_dir):
    # Construct full file path
    file_path = os.path.join(input_dir, file_name)
    
    # Check if it is a file
    if os.path.isfile(file_path):
        # Apply the function to shift the "Load" column up
        shift_load_column_up(file_path)
```
