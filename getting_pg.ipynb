```python
import pandapower as pp
import pandas as pd
import numpy as np
from pandapower.timeseries.output_writer import OutputWriter


# Install pandapower in a new environment by pip install pandapower==2.13.1
net = pp.from_pickle('C:/Users/bjorn/OneDrive/Documents/TU Delft/BEP/TUD Library/bap/aggregated_grid_2018_with_generators_loads_costs_controllers.p')
out_path = "C:/Users/bjorn/OneDrive/Documents/TU Delft/BEP/TUD Library/bap/test_results"
ow = OutputWriter(net, output_path=out_path, output_file_type=".csv", csv_separator=",")
# ow.remove_log_variable("res_gen", "vm_pu")
```

```python
# CHANGING SGEN CONTROLLERS

for controller in net.controller.object:
    if controller.element == 'sgen':
        controller.variable = 'p_mw'
```

```python
# TIMESTEP-WISE SIMULATION TO GET SIMULATED VALUES


# Single time values
def get_simulated_values(t):
    for controller in net.controller.object:
        controller.time_step(t, net)

    pp.runpp(net)
```

```python
# GETTING FROM SIMULATION

def get_from_simulation(res_csv, number_components):
    # Join timeseries dataframes in a dataframe
    bus_timesteps = pd.concat([res_csv[f"{g}"] for g in number_components], axis=1)

    return bus_timesteps
```

```python
# GETTING GENERATION Q

total_bus_numbers = net.bus.index

def get_aggregated_x(res_x, bus_gens, get_from):
    # res_x = csv for get_from_csv or start controller number for get_from_controller
    gen_busses_x = [] # List of timestep-wise gen x dataframes for each bus

    for bus in bus_gens:
        # Join timeseries dataframes (q_mvar, vm_pu and va_degree) at the same bus in a dataframe
        bus_timesteps = get_from(res_x, bus)

        # Add timestep-wise values together
        bus_timesteps_total = bus_timesteps.sum(axis=1)
        
        gen_busses_x.append(bus_timesteps_total)

    gen_x = pd.concat(gen_busses_x, axis=1) # Put busses together in one df (rows = timesteps, columns = buses)

    gen_x.columns = bus_gens.index # Column names to bus numbers

    gen_x = gen_x.reindex(range(total_bus_numbers[0], total_bus_numbers[-1] + 1), axis=1, fill_value=0) # Fill missing bus numbers with q_mvar = 0
    
    return gen_x


def get_from_controller(start_controller_number, number_components):
    # Join load timeseries dataframes at the same bus in a dataframe
    bus_timesteps = pd.concat([net.controller.object[start_controller_number + g].data_source.to_dict()['df'].iloc[:,0] for g in number_components], axis=1)

    return bus_timesteps



# gen_busses_q_mvar = [] 
# gen_busses_vm_pu = []
# gen_busses_va_degree = []



# for bus in bus_gens:
#     # Join timeseries dataframes (q_mvar, vm_pu and va_degree) at the same bus in a dataframe
#     bus_timesteps = pd.concat([res_q_mvar[f"{g}"] for g in bus], axis=1)
#     bus_timesteps_vm = pd.concat([res_vm_pu[f"{vm}"] for vm in bus], axis=1)
#     bus_timesteps_va = pd.concat([res_va_degree[f"{va}"] for va in bus], axis=1)

#     # Add timestep-wise values together
#     bus_timesteps_total = bus_timesteps.sum(axis=1)
#     bus_timesteps_total_vm = bus_timesteps_vm.sum(axis=1)
#     bus_timesteps_total_va = bus_timesteps_va.sum(axis=1)

#     gen_busses_x.append(bus_timesteps_total)
#     gen_busses_vm_pu.append(bus_timesteps_total_vm)    
#     gen_busses_va_degree.append(bus_timesteps_total_va)

# gen_q_mvar = pd.concat(gen_busses_x, axis=1) # Put busses together in one df (rows = timesteps, columns = buses)
# gen_vm_pu = pd.concat(gen_busses_vm_pu, axis=1) # Put busses together in one df (rows = timesteps, columns = buses)
# gen_va_degree = pd.concat(gen_busses_va_degree, axis=1) # Put busses together in one df (rows = timesteps, columns = buses)

# gen_q_mvar.columns = bus_gens.index # Column names to bus numbers
# gen_vm_pu.columns = bus_gens.index # Column names to bus numbers
# gen_va_degree.columns = bus_gens.index # Column names to bus numbers

# gen_q_mvar = gen_q_mvar.reindex(range(total_bus_numbers[0], total_bus_numbers[-1] + 1), axis=1, fill_value=0) # Fill missing bus numbers with q_mvar = 0
# gen_vm_pu = gen_vm_pu.reindex(range(total_bus_numbers[0], total_bus_numbers[-1] + 1), axis=1, fill_value=0) # Fill missing bus numbers with q_mvar = 0
# gen_va_degree = gen_va_degree.reindex(range(total_bus_numbers[0], total_bus_numbers[-1] + 1), axis=1, fill_value=0) # Fill missing bus numbers with q_mvar = 0

# gen_vm_pu.head(100)    
```

```python
# GET GENERATION P_MW (0 and 43 same)

gen_grouped = net.gen.loc[net.gen["in_service"] == True].groupby("bus") # Drop unused generators
gen_p_mw = gen_grouped["p_mw"].sum() # Add generations by bus
gen_p_mw = gen_p_mw.reindex(range(total_bus_numbers[0], total_bus_numbers[-1] + 1), fill_value=0) # Fill missing buses with 0

# gen_p_mw.head(50)
```

```python
# GET GEN Q_MVAR (0 and 43 same)

res_gen_q_mvar = pd.read_csv(out_path + "/res_gen/q_mvar.csv") # Read from simulation result
bus_gens = gen_grouped.apply(lambda x: x.index.tolist()) # Group generators by bus

gen_q_mvar = get_aggregated_x(res_gen_q_mvar, bus_gens, get_from_simulation) # Get bus-aggregated generation

# gen_q_mvar.head(50)
```

```python
# GET LOAD P_MW (0 and 43 same)

bus_loads = net.load.groupby("bus").apply(lambda x: x.index.tolist()) # Group loads by bus

load = get_aggregated_x(0, bus_loads, get_from_controller)

# load.head(50)
```

```python
# GET SGEN (0 and 43 same)

bus_sgens = net.sgen.loc[net.sgen["in_service"] == True].reset_index(drop=True) # Drop unused sgens
bus_sgens = bus_sgens.groupby("bus").apply(lambda x: x.index.tolist()) # Group sgens by bus

sgen = get_aggregated_x(132, bus_sgens, get_from_controller) # Sgen controller starts at 132

# sgen.head(50)
```

```python
# SWAPPING SLACK BUS

def swap_slack(df, slack_number): # Switch slack bus position
    df["0"], df[f"{slack_number}"] = df[f"{slack_number}"], df["0"]

slack_bus = 43
```

```python
# GET BUS ANGLE AND VM_PU

vm_pu = pd.read_csv(out_path + "/res_bus/vm_pu.csv")
swap_slack(vm_pu, slack_bus) 

# vm_pu.head(50)

va_degree = pd.read_csv(out_path + "/res_bus/va_degree.csv")
swap_slack(va_degree, slack_bus)

# va_degree.head(50)
```

```python
# GET SLACK BUS P,Q

res_bus_p_mw = pd.read_csv(out_path + "/res_bus/p_mw.csv")
res_bus_q_mvar = pd.read_csv(out_path + "/res_bus/q_mvar.csv")

slack_p_mw = res_bus_p_mw.iloc[:, 35:]
slack_q_mvar = res_bus_q_mvar.iloc[:, 35:]

slack_p_mw.rename({f"{slack_bus}": "0"})
slack_q_mvar.rename({f"{slack_bus}": "0"})
```

```python
# UPDATE SLACK GEN/LOAD P_MW AND Q_MVAR
```

```python
# GETTING ALL DATA FOR ONE TIMESTEP

zeros = [0 for _ in range(len(total_bus_numbers))]

slack_weight = [0 if i not in range(35, 45) else 1 for i in range(len(total_bus_numbers))] # Slack weights for external grids
slack_weight[0], slack_weight[43] = slack_weight[43], slack_weight[0]

def get_timestep(t):
    total_gen = gen_p_mw.add(sgen.iloc[t]) # sgen + gen
    typ = np.where(total_gen >= load.iloc[t], 2, 3)  # 2 if gen bus, 3 if load bus
    typ[0] = 1 # Bus 0 is slack



    timestep = pd.DataFrame(data={"type": typ, "vm_pu": vm_pu.iloc[t].values, "gen_p_mw": gen_p_mw.values, "gen_q_mvar": gen_q_mvar.iloc[t].values, 
                             "load_p_mw": load.iloc[t].values, "load_q_mvar": zeros, "sgen_p_mw": sgen.iloc[t].values, "sgen_q_mvar": zeros,
                             "slack_weight": slack_weight})
    
    return timestep

# gen.to_csv("C:/Users/bjorn/OneDrive/Documents/TU Delft/BEP/TUD Library/bap/CSV_files/gen.csv")

get_timestep(0).head(50)
```

```python
# CREATING THE CSV WITH ALL TIMESTEPS

def create_csv(t_init, t_duration):
    result = pd.concat([get_timestep(t) for t in range(t_init, t_init + t_duration)], axis=0)
    result.index.name = "Bus"

    result.to_csv(out_path + "/final_csv.csv")
```

```python
# GETTING LOAD

bus_loads = net.load.groupby("bus").apply(lambda x: x.index.tolist())

load_busses = [] # List of timestep-wise load dataframes for each bus

for bus in bus_loads:
    # Join load timeseries dataframes at the same bus in a dataframe
    bus_timesteps = pd.concat([net.controller.object[l].data_source.to_dict()['df'].iloc[:,0] for l in bus], axis=1)
    
    # Add timestep-wise values together
    bus_timesteps_total = bus_timesteps.sum(axis=1)
    load_busses.append(bus_timesteps_total)

load = pd.concat(load_busses, axis=1) # Put busses together in one df
load.columns = bus_loads.index # Column names to bus numbers

load = load.reindex(range(total_bus_numbers[0], total_bus_numbers[-1] + 1), axis=1, fill_value=0) # Fill missing values with 0

load.head(50)

# load.to_csv("C:/Users/bjorn/OneDrive/Documents/TU Delft/BEP/TUD Library/bap/CSV_files/load.csv")
```
